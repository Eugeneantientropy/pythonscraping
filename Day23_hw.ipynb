### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

news_data = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    category = d.find('em').text.strip()  # 抓取新聞類別
    date = d.find(class_="date").text.strip()
    title = d.find_all('a')[-1].text.strip()
    link = d.find_all('a')[-1]['href']

    if category not in news_data:
        news_data[category] = []  # 建立新類別的 key

    news_data[category].append({'date': date, 'title': title, 'link': link})

# 測試印出新聞類別與內容
for category, articles in news_data.items():
    print(f"\n分類：{category}")
    for article in articles[:3]:  # 只顯示前 3 則新聞
        print(f"- {article['date']} | {article['title']} ({article['link']})")
```

### ③ 哪一個類別的文章最多

```python
import requests
from bs4 import BeautifulSoup
from collections import Counter

# 抓取網頁內容
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

# 記錄新聞類別的次數
category_count = Counter()

# 解析新聞分類
for d in soup.find(class_="part_list_2").find_all('h3'):
    category = d.find('em').text.strip()  # 抓取新聞類別
    category_count[category] += 1  # 計算次數

# 找出文章最多的類別
most_common_category = category_count.most_common(1)

most_common_category
```

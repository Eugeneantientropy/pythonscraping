```python
import time
import random
from tqdm import tqdm
from selenium.webdriver.common.by import By

def fetch_prod_list(cate_name, url, driver):
    if url:
        driver.get(url)

    # æ¨¡æ“¬ç”¨æˆ¶éš¨æ©Ÿç­‰å¾…ï¼Œé¿å…è¢«åçˆ¬
    time.sleep(random.uniform(3, 5))

    class_list = []
    
    # æŠ“å–æ‰€æœ‰å•†å“çš„å€å¡Š
    class_cards = driver.find_elements(By.XPATH, '//div[@class="sc-NGAnU lmMDMy"]')

    for prod in tqdm(class_cards):
        try:
            # å•†å“åç¨±
            name = prod.find_element(By.TAG_NAME, "span").text

            # å•†å“åƒ¹æ ¼
            try:
                price = prod.find_element(By.XPATH, './/following-sibling::div[@class="sc-hshbuu isBEJR"]/span[2]').text
            except:
                price = "åƒ¹æ ¼æœªæä¾›"

            # å•†å“é€£çµ
            link = prod.find_element(By.XPATH, './ancestor::a').get_attribute("href")

            class_list.append((name, price, link))
        except Exception as e:
            print(f"ç™¼ç”ŸéŒ¯èª¤: {e}")

    return class_list
```

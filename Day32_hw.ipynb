### 獲取所有小類別頁面連結
> Day29/Day30 內容，可自行載入儲存的爬取表格，也可以再爬一次

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# ✅ 設定 Chrome 選項
options = Options()
options.add_argument("--headless")  # 無頭模式
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")  
options.add_argument("--disable-blink-features=AutomationControlled")  # 防止反爬

# ✅ 啟動 WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# ✅ 目標網址
base_url = 'https://www.cupoy.com/'
driver.get(base_url)

# ✅ 等待網頁完全加載 (最多等待 15 秒)
WebDriverWait(driver, 15).until(
    EC.presence_of_element_located((By.CLASS_NAME, "eFgyzl"))
)

# ✅ 額外等待幾秒，確保內容載入
time.sleep(3)

# ✅ 嘗試獲取分類名稱
categories = driver.find_elements(By.XPATH, '//div[@class="sc-NGAnU lmMDMy"]/span')

data = []

# ✅ 印出分類名稱
for category in categories:
    data.append(category.text)
    print(category.text)


# ✅ 關閉瀏覽器
driver.quit()
```

### 取得課程資訊

```python
import time
import random
from tqdm import tqdm
from selenium.webdriver.common.by import By

def fetch_prod_list(cate_name, url, driver):
    if url:
        driver.get(url)

    # 模擬用戶隨機等待，避免被反爬
    time.sleep(random.uniform(3, 5))

    class_list = []
    
    # 抓取所有商品的區塊
    class_cards = driver.find_elements(By.XPATH, '//div[@class="sc-NGAnU lmMDMy"]')

    for prod in tqdm(class_cards):
        try:
            # 商品名稱
            name = prod.find_element(By.TAG_NAME, "span").text

            # 商品價格
            try:
                price = prod.find_element(By.XPATH, './/following-sibling::div[@class="sc-hshbuu isBEJR"]/span[2]').text
            except:
                price = "價格未提供"

            # 商品連結
            link = prod.find_element(By.XPATH, './ancestor::a').get_attribute("href")

            class_list.append((name, price, link))
        except Exception as e:
            print(f"發生錯誤: {e}")

    return class_list
```

### 模擬「點擊」行為達成換頁，並爬取每一頁商品資訊

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# ✅ 設定 Chrome 選項
options = Options()
options.add_argument("--headless")  # 無頭模式
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")  
options.add_argument("--disable-blink-features=AutomationControlled")  # 防止反爬

# ✅ 啟動 WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# ✅ 目標網址
base_url = 'https://www.cupoy.com/'
driver.get(base_url)

# ✅ 等待網頁完全加載
WebDriverWait(driver, 15).until(
    EC.presence_of_element_located((By.CLASS_NAME, "eFgyzl"))
)

data = []

# ✅ 取得課程數量
num_courses = len(driver.find_elements(By.XPATH, '//a[contains(@class, "eFgyzl")]'))

# ✅ 遍歷課程並點擊進入詳細頁面
for i in range(num_courses):
    try:
        # ✅ 重新獲取課程清單
        courses = driver.find_elements(By.XPATH, '//a[contains(@class, "eFgyzl")]')
        course = courses[i]  # 重新獲取課程元素
        course_title = course.get_attribute("title")  # 課程標題
        course_url = course.get_attribute("href")  # 課程 URL

        print(f"🔎 正在點擊課程 {i+1}: {course_title}")

        # ✅ 點擊課程連結（避免 Stale Element）
        driver.execute_script("arguments[0].scrollIntoView();", course)  # 滾動到可視範圍
        driver.execute_script("arguments[0].click();", course)  # 使用 JavaScript 點擊

        # ✅ 等待新頁面加載
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.TAG_NAME, "h1"))
        )

        # ✅ 擷取課程標題
        title = driver.find_element(By.TAG_NAME, "h1").text

        # ✅ 擷取價格（可能不同 class，嘗試不同方式）
        try:
            price = driver.find_element(By.XPATH, '//div[contains(@class, "sc-SxrYz")]//p').text
        except:
            price = "價格未提供"

        # ✅ 擷取課程描述
        try:
            description = driver.find_element(By.XPATH, '//p[@title]').text
        except:
            description = "無課程描述"

        # ✅ 擷取課程亮點
        try:
            highlights = [li.text for li in driver.find_elements(By.XPATH, '//h3[contains(text(), "課程亮點")]/following-sibling::div//li')]
        except:
            highlights = []

        # ✅ 擷取目標受眾
        try:
            audience = [h2.text for h2 in driver.find_elements(By.XPATH, '//h3[contains(text(), "目標受眾")]/following-sibling::div//h2')]
        except:
            audience = []

        # ✅ 擷取學習目標
        try:
            learning_goals = [li.text for li in driver.find_elements(By.XPATH, '//h3[contains(text(), "學習目標")]/following-sibling::div//li')]
        except:
            learning_goals = []

        # ✅ 擷取必學理由
        try:
            reasons = [li.text for li in driver.find_elements(By.XPATH, '//h3[contains(text(), "必學的理由")]/following-sibling::div//li')]
        except:
            reasons = []

        # ✅ 加入 `data`
        data.append({
            "course_title": title,
            "course_url": course_url,
            "price": price,
            "description": description,
            "highlights": highlights,
            "audience": audience,
            "learning_goals": learning_goals,
            "reasons": reasons
        })

        print(f"📌 {title} 擷取完成！")

        # ✅ 返回主頁，確保頁面重新載入
        driver.back()
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "eFgyzl"))
        )
        time.sleep(2)  # 讓頁面確保完全載入

    except Exception as e:
        print(f"❌ 點擊課程 {i+1} 失敗: {e}")

# ✅ 關閉瀏覽器
driver.quit()

# ✅ 印出所有擷取到的課程資訊
print("✅ 擷取完成，課程列表:")
for d in data:
    print(d)
```
